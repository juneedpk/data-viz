```python
import pandas as pd
from numpy import median
import matplotlib.pyplot as plt
import seaborn as sns


kashti = sns.load_dataset('titanic')

sns.barplot(x='class',y='fare',linewidth=5,errcolor='0.3',edgecolor='0.2',data=kashti,facecolor=(1,0.6,0.2,1))
plt.xlabel('comparison of sex',size=24)
plt.ylabel('what is age',size=20)
plt.title("graph of titanic",size=14)
plt.show()

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


kashti = sns.load_dataset('titanic')

# kashti.head()

sns.boxplot(x='sex',y='age',data=kashti,dodge=True,saturation=0.3,palette='dark',showmeans=True,meanprops={'marker':'*','markersize':'24','markeredgecolor':'blue'})

plt.xlabel('comparison of sex',size=24)
plt.ylabel('what is age',size=20)
plt.title("graph of titanic",size=14,weight='bold')

```
